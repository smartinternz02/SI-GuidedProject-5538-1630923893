```python
import pandas
import matplotlib
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.species.unique()
```

```python
from matplotlib import pyplot as plt
```

```python
x=iris.species
y=iris.petal_width
plt.bar(x,y)
plt.show()
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width)
plt.show()
```

```python
sns.set_style('dark')
sns.kdeplot(iris.loc[(iris['species']=='setosa'),
                     'sepal_length'],shade=True,color='b',Label='Setosa')
sns.kdeplot(iris.loc[(iris['species']=='virginica'),
                     'sepal_length'],shade=True,color='r',Label='virginica')
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
```

```python
plt.hist(x=iris.petal_length,color='g',bins=20)
plt.title('petal_length')
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(iris.petal_width)
```

```python
sns.set_style('whitegrid')
sns.relplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

```python
iris['species'].value_counts().plot.pie()
plt.title("Iris Species ")
plt.show()
```

```python
sns.set_style('dark')
sns.kdeplot(iris.loc[(iris['species']=='setosa'),
                     'sepal_length'],shade=True,color='b',Label='Setosa')
```

```python
x=iris.species
y=iris.sepal_width
plt.plot(x,y)
```

```python
y=iris.species
x1=iris.sepal_length
x2=iris.sepal_width
fig=plt.figure()
ax=fig.add_axes([0,0,2,1])
l1=ax.plot(x1,y,'g-')
l2=ax.plot(x2,y,'r--')
ax.legend(labels=('Line1','Line2'),loc='upper right')
ax.set_title("add axes")
ax.set_xlabel('xaxis')
ax.set_ylabel('yaxis')
plt.show()
```
